# 2.Handle missing values using mean/median/imputation

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

Median is used by most data analyst, as the values in the Age column are not symterical and rather skewed, so mean is the not best option.  

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

# 4.Normalize/standardize the numerical features.

Numerical features can have very different scales with may affect the model while training.

It may consider some features more important than others, even if there is no relevance.

# Method 1 : Standardization

Formula = (Value - mean) / Standard Deviation

After Stanadardization:  mean = 0 , Standard DEviation = 1



```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[["Age","Fare"]] = scaler.fit_transform(df[["Age","Fare"]])
df.head()
```

# Method 2: Normalization

Formula: x' = (x - x_min) / (x_max - x_min)

After Normalization: min = 0 , max = 1  

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[["Age", "Fare"]] = scaler.fit_transform(df[["Age", "Fare"]])
df.head()
```

| Standardization                  | Normalization                      |
| -------------------------------- | ---------------------------------- |
| Mean = 0, Std = 1                | Range = 0 to 1                     |
| Not affected as much by outliers | Sensitive to outliers              |
| Most commonly used               | Used when a fixed range is desired |

# 5.1 Identifying Outliers using IQR

IQR = Interquartile Range

Formula = Q3 - Q1

where, Q1 = 25th percentile , Q3 = 75th percentile


```python
q1 = df["Fare"].quantile(0.25)
q3 = df["Fare"].quantile(0.75)
IQR = q3 - q1

lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR
```

# 5.2 Remove Outliers

```python
df = df[
    (df["Fare"] >= lower_bound) &
    (df["Fare"] <= upper_bound)

]
```
