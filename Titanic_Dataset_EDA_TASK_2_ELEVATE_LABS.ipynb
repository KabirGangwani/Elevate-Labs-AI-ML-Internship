# Step 1: Generate Summary Statistics

```python
import pandas as pd
df = pd.read_csv(
    "https://raw.githubusercontent.com/KabirGangwani/Titanic-Data-Preprocessing/main/Titanic-Dataset.csv"
)
```

```python
df.head()
```

# Step 1: Generate Summary Statistics

In this step, we generate summary statistics to understand the distribution and central tendency of numerical features such as Age and Fare.

```python
df.describe()
```

# Step 2: Histograms for Numerical Features

Histograms help visualize the distribution of numerical variables and identify skewness, spread, and potential outliers.

```python
import matplotlib.pyplot as plt

df.hist(figsize=(12,8))
plt.show()
```

# Step 3: Boxplots for Numerical Features

Boxplots help identify outliers and visualize the spread of numerical data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = df.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

# Step 4: Correlation Matrix

A correlation matrix helps understand relationships between numerical features.
Correlation values range from -1 to 1.

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Matrix")
plt.show()
```

# Step 5: Pairplot

Pairplots help visualize pairwise relationships between numerical variables.

```python
sns.pairplot(
    df[['Age','Fare','Pclass','Survived']]
)

plt.show()
```

# Step 6: Survival Count Analysis

This visualization shows the number of passengers who survived and did not survive.

```python
sns.countplot(
    x='Survived',
    data=df
)

plt.title("Survival Distribution")
plt.show()
```

# Step 7: Survival Based on Gender

This visualization helps identify whether survival rates differed between male and female passengers.

```python
sns.countplot(
    x='Sex',
    hue='Survived',
    data=df
)

plt.title("Survival by Gender")
plt.show()
```

# Step 8: Survival Based on Passenger Class

This analysis examines the relationship between passenger class and survival.

```python
sns.countplot(
    x='Pclass',
    hue='Survived',
    data=df
)

plt.title("Survival by Passenger Class")
plt.show()
```

# Step 9: Key Insights

Observations from the Exploratory Data Analysis:

1. Age and Fare distributions contain outliers.
2. Female passengers had a higher survival rate.
3. First-class passengers were more likely to survive.
4. Fare is positively correlated with survival.
5. Several numerical features exhibit skewed distributions.
